```python
# Ejercicio N° 2.2.
# 1: Definimos el enumerado de tipos de planeta.
from enum import Enum
class TipoPlaneta(Enum):
    GASEOSO = 1
    TERRESTRE = 2
    ENANO = 3

# 2: Definimos la clase Planeta.
class Planeta:
    def __init__(self, nombre=None, cantSatelites=0, masa=0.0, volumen=0.0, diametro=0, distanciaSol=0, tipo=TipoPlaneta.TERRESTRE, observable=False, periodoOrbital=0.0, periodoRotacion=0.0):
        self.nombre = nombre
        self.cantSatelites = cantSatelites
        self.masa = masa
        self.volumen = volumen
        self.diametro = diametro
        self.distanciaSol = distanciaSol
        self.tipo = tipo
        self.observable = observable
        self.periodoOrbital = periodoOrbital
        self.periodoRotacion = periodoRotacion

# Método para mostrar todos los datos
    def mostrarDatos(self):
        print("Nombre:", self.nombre)
        print("Satélites:", self.cantSatelites)
        print("Masa:", self.masa)
        print("Volumen:", self.volumen)
        print("Diámetro:", self.diametro)
        print("Distancia al Sol:", self.distanciaSol)
        print("Tipo:", self.tipo.name)
        print("Observable:", self.observable)
        print("Periodo orbital (años):", self.periodoOrbital)
        print("Periodo de rotación (días):", self.periodoRotacion)

# Método para calcular densidad (masa / volumen)
    def calcularDensidad(self):
        if self.volumen > 0:
            return self.masa / self.volumen
        return 0

# Método para saber si es planeta exterior.
    def esPlanetaExterior(self):
        return self.distanciaSol > 3.4

# 3: Solicitar datos al usuario.

print("Digite los datos del primer planeta")
nombre = input("Nombre: ")
satelites = int(input("Cantidad de satélites: "))
masa = float(input("Masa (kg): "))
volumen = float(input("Volumen (km^3): "))
diametro = int(input("Diámetro (km): "))
distancia = int(input("Distancia al Sol (UA): "))
tipo = int(input("Tipo de planeta (1=GASEOSO, 2=TERRESTRE, 3=ENANO): "))
observable = input("¿Es observable? (s/n): ") == "s"
periodoOrbital = float(input("Periodo orbital (años): "))
periodoRotacion = float(input("Periodo de rotación (días): "))

# 4: Crear el objeto planeta con los datos ingresados.

p1 = Planeta(nombre, satelites, masa, volumen, diametro, distancia, TipoPlaneta(tipo), observable, periodoOrbital, periodoRotacion)

# 5: Mostrar resultados.
print("\nDatos del planeta ingresado:")
p1.mostrarDatos()
print("Densidad:", p1.calcularDensidad())
print("¿Es planeta exterior?:", p1.esPlanetaExterior())
```

```python
# Ejercicio N° 2.3.
# 1: Definimos enumeraciones.
from enum import Enum

class Tipo_Combustible(Enum):
    Gasolina = 1
    Bioetanol = 2
    Diesel = 3
    Biodiesel = 4
    Gas_Natural = 5

class Tipo_Automovil(Enum):
    Ciudad = 1
    Subcompacto = 2
    Compacto = 3
    Familiar = 4
    Ejecutivo = 5
    Suv = 6

class Tipo_Color(Enum):
    Blanco = 1
    Negro = 2
    Rojo = 3
    Naranja = 4
    Amarillo = 5
    Verde = 6
    Azul = 7
    Violeta = 8
# 2: Clase Automóvil.
class Automovil:
    def __init__(self, Marca, Modelo, Motor, Tipo_Combustible, Tipo_Auto,
                 Numero_Puertas, Cantidad_Asientos, Velocidad_Maxima, Color):
        self.Marca = Marca
        self.Modelo = Modelo
        self.Motor = Motor
        self.Tipo_Combustible = Tipo_Combustible
        self.Tipo_Auto = Tipo_Auto
        self.Numero_Puertas = Numero_Puertas
        self.Cantidad_Asientos = Cantidad_Asientos
        self.Velocidad_Maxima = Velocidad_Maxima
        self.Color = Color
        self.Velocidad_Actual = 0

# Métodos de comportamiento
    def Acelerar(self, Incremento_Velocidad):
        if self.Velocidad_Actual + Incremento_Velocidad > self.Velocidad_Maxima:
            self.Velocidad_Actual = self.Velocidad_Maxima
            print(f"No se puede superar la velocidad máxima. Velocidad actual: {self.Velocidad_Actual} km/h")
        else:
            self.Velocidad_Actual += Incremento_Velocidad
            print(f"Velocidad actual: {self.Velocidad_Actual} km/h")

    def Desacelerar(self, Desaceleracion):
        if self.Velocidad_Actual - Desaceleracion < 0:
            self.Velocidad_Actual = 0
            print("No se puede ir a velocidad negativa. Velocidad actual: 0 km/h")
        else:
            self.Velocidad_Actual -= Desaceleracion
            print(f"Velocidad actual: {self.Velocidad_Actual} km/h")

    def Frenar(self):
        self.Velocidad_Actual = 0
        print("El automóvil se ha detenido. Velocidad actual: 0 km/h")

    def Calcular_Tiempo_Llegada(self, Distancia):
        if self.Velocidad_Actual == 0:
            print("No se puede calcular tiempo de llegada porque la velocidad es 0.")
            return None
        return Distancia / self.Velocidad_Actual

    def Datos_(self):
        print("\nDatos del Automóvil\n")
        print(f"Marca: {self.Marca}")
        print(f"Modelo: {self.Modelo}")
        print(f"Motor: {self.Motor} L")
        print(f"Combustible: {self.Tipo_Combustible.name}")
        print(f"Tipo: {self.Tipo_Auto.name}")
        print(f"Puertas: {self.Numero_Puertas}")
        print(f"Asientos: {self.Cantidad_Asientos}")
        print(f"Velocidad Máxima: {self.Velocidad_Maxima} km/h")
        print(f"Color: {self.Color.name}")
        print(f"Velocidad Actual: {self.Velocidad_Actual} km/h")

# 3: Funciones auxiliares.
def Elegir_Enum(Enum_Class, Mensaje):
    print(f"\nOpciones disponibles en {Enum_Class.__name__}:")
    for Dato in Enum_Class:
        print(f"{Dato.value}. {Dato.name}")

    while True:
        try:
            Entrada = int(input(Mensaje))
            if Entrada in [E.value for E in Enum_Class]:
                return Enum_Class(Entrada)
            else:
                print("Opción inválida, intenta de nuevo.")
        except ValueError:
            print("Debes ingresar un número válido.")

def Validar_Entrada(mensaje, tipo, Restriccion):
    while True:
        entrada = input(mensaje)

        try:
            if tipo == str and entrada.lower().isalpha():
                return entrada.strip()
            elif tipo == int and entrada.isdigit():
                if Restriccion == len(entrada):
                    return entrada
                elif Restriccion == len(entrada):
                    return entrada
                elif Restriccion == len(entrada):
                    return entrada
                elif Restriccion == len(entrada):
                    return entrada
                else:
                    print("Cantidad de digitos no validos.\n")
            else:
                print("El sistema no admite este tipo de dato. Inténtalo de nuevo.")
        except ValueError:
            print("Error. Debes ingresar un valor adecuado")

# 4: Programa principal
if __name__ == "__main__":
    print("¿Datos del Automóvil?\n")

    Marca = Validar_Entrada("Marca del Automóvil:\n",str,False)
    Modelo = Validar_Entrada("Modelo (año de fabricación):\n ",int,4)
    Motor = Validar_Entrada("Motor (cilindraje en litros): \n",int,1)

    Tipo_Combustible_Seleccionado = Elegir_Enum(Tipo_Combustible, "Seleccione el tipo de Combustible: ")
    Tipo_Auto_Seleccionado = Elegir_Enum(Tipo_Automovil, "Seleccione el tipo de Automóvil: ")
    Color_Seleccionado = Elegir_Enum(Tipo_Color, "Seleccione el color del Automóvil: ")

    Numero_Puertas = Validar_Entrada("Número de Puertas: ",int,1)
    Cantidad_Asientos = Validar_Entrada("Cantidad de Asientos: ",int,1)
    Velocidad_Maxima = Validar_Entrada("Velocidad máxima (km/h): ",int,3)

 # Creamos objeto Automovil
    Auto = Automovil(Marca, Modelo, Motor, Tipo_Combustible_Seleccionado, Tipo_Auto_Seleccionado,
                     Numero_Puertas, Cantidad_Asientos, Velocidad_Maxima, Color_Seleccionado)
# Mostrar datos.
    Auto.Datos_()
```

```python
# Ejercicio N° 2.4.
# 1: Definir la clase Círculo.
import math
class Circulo:
    def __init__(self, radio):
        self.radio = radio

    def CalcularArea(self):
        return math.pi * math.pow(self.radio, 2)

    def CalcularPerimetro(self):
        return 2 * math.pi * self.radio

# 2: Definir la clase Rectángulo.
class Rectangulo:
    def __init__(self, Base, Altura):
        self.Base = Base
        self.Altura = Altura

    def CalcularArea(self):
        return self.Base * self.Altura

    def CalcularPerimetro(self):
        return 2 * (self.Base + self.Altura)

# 3: Definir la clase Cuadrado.
class Cuadrado:
    def __init__(self, Lado):
        self.Lado = Lado

    def CalcularArea(self):
        return self.Lado ** 2

    def CalcularPerimetro(self):
        return 4 * self.Lado

# 4: Definir la clase Triángulo Rectángulo.
class TrianguloRectangulo:
    def __init__(self, Base, Altura):
        self.Base = Base
        self.Altura = Altura

    def CalcularArea(self):
        return (self.Base * self.Altura / 2)

    def CalcularHipotenusa(self):
        return math.pow(self.Base*self.Base + self.Altura*self.Altura, 0.5)

    def CalcularPerimetro(self):
        return (self.Base + self.Altura + self.CalcularHipotenusa())

    def DeterminarTipoTriangulo(self):
        if self.Base == self.Altura and self.Base == self.CalcularHipotenusa() and self.Altura == self.CalcularHipotenusa():
            print("Es un triángulo equilátero")
        elif self.Base != self.Altura and self.Base != self.CalcularHipotenusa() and self.Altura != self.CalcularHipotenusa():
            print("Es un triángulo escaleno")
        else:
            print("Es un triángulo isósceles")

# 5: Crear una función de prueba para todas las figuras
def PruebaFiguras():
    figura1 = Circulo(2)
    figura2 = Rectangulo(1, 2)
    figura3 = Cuadrado(3)
    figura4 = TrianguloRectangulo(3, 5)

    print("Infome general de las figuras:\n")

    print("El área del círculo es =", figura1.CalcularArea())
    print("El perímetro del círculo es =", figura1.CalcularPerimetro())

    print("\nEl área del rectángulo es =", figura2.CalcularArea())
    print("El perímetro del rectángulo es =", figura2.CalcularPerimetro())

    print("\nEl área del cuadrado es =", figura3.CalcularArea())
    print("El perímetro del cuadrado es =", figura3.CalcularPerimetro())

    print("\nEl área del triángulo rectangulo es =", figura4.CalcularArea())
    print("El perímetro del triángulo rectangulo es =", figura4.CalcularPerimetro())
    figura4.DeterminarTipoTriangulo()

#6: Ejecutar el programa principal
if __name__ == "__main__":
    PruebaFiguras()
```

```python
#Actividaad N° 2.5
# 1. Definición del valor enumerado para el Tipo de Cuenta
from enum import Enum
class CuentaTipo(Enum):
    AHORROS = "AHORROS"
    CORRIENTE = "CORRIENTE"
    def __str__(self):
        return self.value
# 2. Definición de la clase CuentaBancaria
class CuentaBancaria:
# Modelamos una cuenta bancaria con funcionalidades básicas
# Constructor
    def __init__(self, nombres_titular: str, apellidos_titular: str, numero_cuenta: int, tipo_cuenta: CuentaTipo, interes_mensual: float = 0.0):

# Iniciamos con los atributos de la cuenta bancaria
        self.nombres_titular = nombres_titular
        self.apellidos_titular = apellidos_titular
        self.numero_cuenta = numero_cuenta
        self.tipo_cuenta = tipo_cuenta
        self.saldo = 0.0
        self.interes_mensual = float(interes_mensual)

# Método para imprimir todos los datos en pantalla
    def imprimir_datos(self) -> None:
        print(f"Nombres del titular = {self.nombres_titular}")
        print(f"Apellidos del titular = {self.apellidos_titular}")
        print(f"Número de cuenta = {self.numero_cuenta}")
        print(f"Tipo de cuenta = {self.tipo_cuenta}")
        print(f"Saldo = ${self.saldo:.2f}")

# Método para consultar el saldo.
    def consultar_saldo(self) -> float:
        print(f"El saldo actual es = ${self.saldo:.2f}")
        return self.saldo

# Método para consultar el saldo.
    def consignar(self, valor: float) -> bool:
        if valor > 0:
            self.saldo += valor
            print(f"Se ha consignado ${valor:.2f} en la cuenta. El nuevo saldo es ${self.saldo:.2f}")
            return True
        else:
            print("El valor a consignar debe ser mayor que cero.")
            return False

# Método para retirar.
    def retirar(self, valor: float) -> bool:
        if valor > 0 and valor <= self.saldo:
            self.saldo -= valor
            print(f"Se ha retirado ${valor:.2f} de la cuenta. El nuevo saldo es ${self.saldo:.2f}")
            return True
        else:
            print("El valor a retirar debe ser mayor que cero y no superar el saldo actual.")
            return False

    def calcular_saldo_con_interes(self, meses: int = 1) -> float:
        if meses < 0:
            raise ValueError("Los meses deben ser >= 0")
        factor = (1 + self.interes_mensual / 100) ** meses
        return self.saldo * factor

    def aplicar_interes(self, meses: int = 1) -> float:
        nuevo_saldo = self.calcular_saldo_con_interes(meses)
        self.saldo = nuevo_saldo
        print(f"Se aplicó interés del {self.interes_mensual:.4g}% mensual por {meses} mes(es). Nuevo saldo: ${self.saldo:.2f}")
        return self.saldo

# 3.Ejecución del programa
if __name__ == "__main__":
    cuenta = CuentaBancaria("Alex", "Hurtado", 123456789, CuentaTipo.AHORROS)
    cuenta.imprimir_datos()
    cuenta.consignar(200000)
    cuenta.consignar(300000)
    cuenta.retirar(400000)
    cuenta.consultar_saldo()

# Ejercicio propuesto: interés mensual (ejemplo)
    cuenta.interes_mensual = 2.5
    cuenta.aplicar_interes(1)
```
